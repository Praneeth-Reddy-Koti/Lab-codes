variable while the other as dependent variable (you may also round this number to integer). After 
the model is created, calculate the mean square error by predicting the values from the model.

```python
independent_variable=new_df['hour']
dependent_variable=new_df['price']
```

```python
independent_variable = independent_variable.dropna()
dependent_variable = dependent_variable.dropna()
```

```python
mean_independent=independent_variable.mean()
mean_dependent=dependent_variable.mean()
```

```python
# For the calculation of theta-1
numerator = float()
denomenator = float()

for i,j in zip(independent_variable,dependent_variable):
   numerator += (float(i-mean_independent)*float(j-mean_dependent))
   denomenator += float((i-mean_independent)**2)
    
print(numerator)
print(denomenator)
```

```python
theta_1=numerator/denomenator
print(theta_1)
```

```python
theta_0=float(mean_dependent) - float(theta_1 * mean_independent)

x=float()
y=theta_1 * x + theta_0
```
